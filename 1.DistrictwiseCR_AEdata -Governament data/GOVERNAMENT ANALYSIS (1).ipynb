# DistrictswiseCR_AEdataf_24-25
This dataset appears to contain information about case registrations and advice-enabled services across different states and
districts in India for the year 2024-2025. It categorizes data based on different parameters such as gender, caste, and the
number of Common Service Centers (CSCs).

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("DistrictswiseCR_AEdataf_24-25.csv")
df
```

Missing values count
Memory usage

```python
display(df.info())
```

```python
display(df.isnull().sum())
```

```python
display(df.describe())
```

Aggregates case counts for different categories (General, OBC, SC).

```python
df_grouped_state = df.groupby("Category")["General","OBC","SC"].sum()
display(df_grouped_state)
```

```python
Visualizes the count of cases in the General category using a bar chart.
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df_grouped_state.index, y=df_grouped_state["General"], palette="viridis")
plt.xticks(rotation=90)
plt.title("General category")
plt.xlabel("category")
plt.ylabel("Count")
plt.show()
```

```python
Displays a heatmap showing case distribution across categories and caste groups.
```

```python
plt.figure(figsize=(8, 5))
heatmap_data = df.groupby("Category")[["General", "OBC", "SC", "ST"]].sum()
sns.heatmap(heatmap_data, cmap="coolwarm", annot=True, fmt="d", linewidths=0.5)
plt.title("Heatmap of Case Distribution Across Categories & Caste Groups")
plt.xlabel("Caste Group")
plt.ylabel("Category")
plt.show()
```

```python
Shows the proportion of male and female cases in a pie chart.
```

```python
df_gender_total = df.groupby("Category")[["Male", "Female"]].sum().sum()
plt.figure(figsize=(6, 6))
plt.pie(
    df_gender_total, 
    labels=["Male", "Female"], 
    autopct="%1.1f%%", 
    colors=["#3498db", "#e74c3c"], 
    startangle=90
)
```

```python
Plots the total number of cases for each state using a line chart.
```

```python
plt.figure(figsize=(12, 5))
statewise_cases = df.groupby("States/UT's")["Total"].sum().sort_values()
statewise_cases.plot(kind="line", marker="o", color="b", linestyle="-", linewidth=2)
plt.title("Total Cases Across States")
plt.xlabel("States/UTs")
plt.ylabel("Total Cases")
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
Displays outliers in case registrations and advice-enabled cases.
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(y=df["Total"], x=df["Category"], palette="pastel")
plt.title("Outlier Analysis: Case Registrations & Advice Enabled")
plt.xlabel("Category")
plt.ylabel("Total Cases")
plt.show()
```

```python
 Bar chart showing the states with the highest number of Common Service Centers (CSCs).
```

```python
df.groupby("States/UT's")["No. of CSCs"].sum().sort_values(ascending=False).head(10).plot(
    kind="bar", color="skyblue", title="Top 10 States by Number of CSCs"
)
```

```python
df["Category"].value_counts().plot(
    kind="pie", autopct="%1.1f%%", startangle=90, title="Category Distribution"
)
```

```python
df.groupby("Districts")["Total"].sum().sort_values(ascending=False).head(10).plot(
    kind="barh", color="teal", title="Top 10 Districts by Total Cases"
)
```

```python
top_states = df.groupby("States/UT's")["Total"].sum().sort_values(ascending=False).head(10).index
df[df["States/UT's"].isin(top_states)].groupby("States/UT's")[["Female", "Male"]].sum().plot(
    kind="bar", stacked=True, title="Gender-wise Distribution in Top 10 States"
)
```

```python
plt.figure(figsize=(12,6))
df_grouped_district = df.groupby("Districts")["Female", "Male"].sum()
df_grouped_district.sort_values("Female", ascending=False).head(10).plot(kind="bar", figsize=(12,6))
plt.title("Top 10 Districts by Female")
plt.xlabel("Districst")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
top_5_states = df_grouped_state.sort_values("General", ascending=False).head(5)
display(top_5_states)
```

```python
plt.figure(figsize=(8,8))
df_grouped_state["General"].plot(kind="pie", autopct="%1.1f%%", cmap="Set3")
plt.title("Percentage of General")
plt.ylabel("")
plt.show()
```

```python
Analyzes the relationship between female and male cases.
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x="General", y="SC", label="Initiated", marker='o')
sns.lineplot(data=df, x="General", y="ST", label="Approved", marker='s')
plt.title("District wise over the year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
df_grouped_district["Gender"] = (df_grouped_district["Female"] / df_grouped_district["Male"]) * 100
display(df_grouped_district[["Gender"]].sort_values("Gender", ascending=False))
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_grouped_district.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Male and Female at District Level")
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.lineplot(x="Female", y="Male", hue="General", data=df, marker="o")
plt.xticks(rotation=90)
plt.title("general of Female and male")
plt.xlabel("Female")
plt.ylabel("Male")
plt.legend(title="Products", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
total_initiated = df["Female"].sum()
total_approved = df["Male"].sum()

approval_accuracy = (total_approved / total_initiated) * 100
print(f"Overall Approval Accuracy: {approval_accuracy:.2f}%")
```

Observations from the Dataset: "DistrictswiseCR_AEdataf_24-25.csv"
The dataset differentiates between cases registered and advice-enabled services in each district.
Some states have more advice-enabled cases than registered cases, indicating people are seeking legal advice before registering cases.
The dataset categorizes cases into General, OBC, SC, and ST categories.
SC & ST communities have a significant number of registered cases, highlighting legal challenges in these communities.
Some states have higher case registration for OBC groups, indicating caste-related legal issues may vary by region.
Certain states/districts have higher case registration rates, indicating high legal activity.
Some regions show low case registration but high advice-enabled cases, suggesting people are seeking guidance rather than taking legal action.
Differences in legal participation across states may be influenced by local policies, awareness campaigns, or socio-economic factors.
